```python
import cv2
import numpy as np
import copy, traceback
# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture('26-3-2022/output4/2022-03-26-10-00.mp4')
W = int(cap.get(3))
H = int(cap.get(4))
video_write = cv2.VideoWriter("id_output4.mp4", cv2.VideoWriter_fourcc('m', 'p', '4', 'v'), 60, (W, H))
# Check if camera opened successfully
if (cap.isOpened()== False): 
      print("Error opening video stream or file")
frame_num = 0

# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frameq
    ret, frame = cap.read()
    copied_frame = copy.copy(frame)
    if ret == True:
    # Display the resulting frame
        my_img_1 = np.zeros((copied_frame.shape[0], copied_frame.shape[1], 1), dtype = "uint8")
        for i in range( len(data['frame'])):
            
            try:
                if data['frame'][i] > frame_num: break
                x,y = data['x'][i], data['y'][i]
                
                my_img_1 = cv2.circle(my_img_1, (x, y), radius=4, color=(255,255,255), thickness=-1)
                #color_id = (int(all_color[int(data['id'][i])][0]), int(all_color[int(data['id'][i])][1]),int(all_color[int(data['id'][i])][2]))
                #color = np.random.randint(0, 255, size=(3, ))
                #color_id = ( int (color [ 0 ]), int (color [ 1 ]), int (color [ 2 ])) 
                #frame = draw_circle(frame, c_x, c_y, (0, 0, 255))
                frame = cv2.line(frame, (data['x'][i], data['y'][i]), (data['x'][i+1], data['y'][i+1]), (255, 0, 0), thickness = 5)
                frame = cv2.line(frame, (0, 885), (1150, 960), (255, 0, 0), thickness = 10)
                frame = cv2.line(frame, (1150, 960), (1920, 920), (255, 0, 0), thickness = 10)
            except:
                    print(traceback.format_exc())

        cv2.imwrite(f'E:/Download/output/out/{frame_num}.jpg', frame)
        #video_write.write(frame)
        
        frame_num += 1
        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
              break

        # Break the loop
    else: break

# When everything done, release the video capture object
cap.release()
video_write.release()
# Closes all the frames
cv2.destroyAllWindows()
```
