```python
import pandas
data=pandas.read_table('26-3-2022/output4/results.txt', 
                           delimiter = ',', header=None, names=['frame', 'x', 'y', 'id'])
```

```python
data
```

```python
import numpy as np
groups = []
ids = data.groupby("id")
for id, group in ids:
    group.reset_index(drop=True, inplace=True)
    #info = (id, group)
    info = (id, group, tuple(np.random.random(size=3) * 256))
    groups.append(info)
```

```python
groups[0]
```

```python
import cv2
import numpy as np
import copy, traceback, time
# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
cap = cv2.VideoCapture('26-3-2022/output4/2022-03-26-10-00.mp4')
W = int(cap.get(3))
H = int(cap.get(4))
video_write = cv2.VideoWriter("26-3-2022/output4/output_green.mp4", cv2.VideoWriter_fourcc('m', 'p', '4', 'v'), 60, (W, H))
# Check if camera opened successfully
if (cap.isOpened()== False): 
      print("Error opening video stream or file")
frame_num = 0

# Read until video is completed
while(cap.isOpened()):
  # Capture frame-by-frame
    tik = time.time()
    ret, frame = cap.read()
    copied_frame = copy.copy(frame)
    
    if ret == True:
        #my_img_1 = np.zeros((copied_frame.shape[0], copied_frame.shape[1], 1), dtype = "uint8")
        
        tik2 = time.time()
        for info in groups:
            id, group, color = info
            #id, group = info
            for i in range(len(group['frame'])-1):
                try:
                    #if group['frame'][i] > frame_num:  break
                    if group['frame'][i] >= (frame_num-20) and group['frame'][i] <= frame_num:
                        x,y = group['x'][i], group['y'][i]
                        #my_img_1 = cv2.circle(my_img_1, (x, y), radius=4, color=(255,255,255), thickness=-1)
                        #frame = cv2.line(frame, (group['x'][i], group['y'][i]), (group['x'][i+1], group['y'][i+1]), (0, 255, 0) , thickness = 5)
                        frame = cv2.line(frame, (group['x'][i], group['y'][i]), (group['x'][i+1], group['y'][i+1]), color , thickness = 5)
                        frame = cv2.line(frame, (0, 885), (1150, 960), (255, 0, 0), thickness = 10)
                        frame = cv2.line(frame, (1150, 960), (1920, 920), (255, 0, 0), thickness = 10)
                except:
                    print(traceback.format_exc())
        tok2 = time.time()
        print('computation took ', tok2-tik2, ' seconds')

        tik1 = time.time()
        video_write.write(frame)
        tok1 = time.time()
        print('wrrite frame took ', tok1-tik1, ' seconds')

        frame_num += 1
        tok = time.time()
        print("finish write frame_num:", frame_num, 'tooks:', tok-tik, 'seconds')
        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
              break
        # Break the loop
    else: 
        break
    
# When everything done, release the video capture object
cap.release()
video_write.release()
# Closes all the frames
cv2.destroyAllWindows()
```
